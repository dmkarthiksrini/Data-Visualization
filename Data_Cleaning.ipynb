```python
import pandas as pd
import numpy as np
import re
```

```python
def extract_base_number(value):
    if pd.isna(value):
        return value
    else:
        match = re.match(r'^(\d+)[+-](\d+)$', str(value))
        if match:
            base = int(match.group(1))
            left = int(match.group(2))
            if '+' in value:
                return base + left
            else:
                return base - left
        else:
            return value
```

```python
datatest = pd.read_csv("Data_Visualization_DataSet.csv",low_memory=False)
```
